```python
import numpy as np
import matplotlib.pyplot as plt
import time 
```

```python
#Bubble Sort
def bubblesort(arr):
  n  = len(arr)
  for i in range(n-1):
    for j in range (n-i-1):
      if arr[j] > arr[j+1]:
        #Swapping the numbers
        arr[j], arr[j+1] = arr[j+1], arr[j]
```

```python
def selectionsort(arr):
  n = len(arr)
  for i in range (n-1):
    min_index = i
    for j in range(i+1, n):
      if arr[j] < arr[min_index]:
        min_index = j 
      arr[i], arr[min_index] = arr[min_index], arr[i]

def quickSort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]  # Choosing the first element as the pivot
        left = [x for x in arr[1:] if x < pivot]
        right = [x for x in arr[1:] if x >= pivot]
        return quickSort(left) + [pivot] + quickSort(right)

```

```python
#Measuring The time
def measuretime(sort, arr):
    start_time = time.time()
    sort(arr)
    end_time = time.time()
    return end_time - start_time
```

```python
#Generating Array and Storing Times per array
array_sizes = [1000, 2000, 3000, 4000, 5000, 6000]
bubble_sort_times = []
selection_sort_times = []
quick_sort_times = []
for size in array_sizes:
    arr = np.random.randint(1, 1000, size).tolist()
    arr_copy = arr[:]
    arr_copy1 = arr[:]
    
    bubble_sort_times.append(measuretime(bubblesort, arr))
    selection_sort_times.append(measuretime(selectionsort, arr_copy))
    quick_sort_times.append(measuretime(quickSort, arr_copy1))

```

```python
plt.figure(figsize = (10, 5))
plt.plot(array_sizes, bubble_sort_times, label = "Bubble", marker = 'o')
plt.plot(array_sizes, selection_sort_times, label = "Selection", marker = 'o')
plt.plot(array_sizes, quick_sort_times, label = "Quick", marker = 'o')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.title('Bubble vs Selection vs Quick')
plt.legend()
plt.grid(True)
plt.show()
```
